```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.colors as mcolors
from matplotlib.lines import Line2D

def plot_test_summary_final_corrected(df):
    plot_df = df.copy()
    
    # ------------------------------------------------------------------
    # Parse 'sid' string: "SIZE x LENGTH" (e.g., "64x4")
    # ------------------------------------------------------------------
    plot_df['codebook_size'] = plot_df['sid'].apply(lambda x: int(x.split('x')[0]))
    plot_df['sid_length'] = plot_df['sid'].apply(lambda x: int(x.split('x')[1]))

    unique_sizes = sorted(plot_df['codebook_size'].unique())
    unique_lengths = sorted(plot_df['sid_length'].unique())
    
    # Mappings
    palette = sns.color_palette("colorblind", n_colors=len(unique_lengths))
    len_to_color = dict(zip(unique_lengths, palette))
    
    size_alphas = np.linspace(0.3, 1.0, len(unique_sizes))
    size_to_alpha = dict(zip(unique_sizes, size_alphas))

    # Setup Grid: [Scatter Plot (1.0)] [Colorbars (0.1)]
    # Decreased width_ratio for colorbars to make them thin
    fig = plt.figure(figsize=(5, 7))
    gs = fig.add_gridspec(1, 2, width_ratios=[1, 0.1], wspace=0.05)
    
    # --- Main Plot (Broken Axis) ---
    gs_main = gs[0].subgridspec(2, 1, height_ratios=[1, 1], hspace=0.1)
    ax_top = fig.add_subplot(gs_main[0])
    ax_bot = fig.add_subplot(gs_main[1], sharex=ax_top)

    def plot_points(target_ax, metric_col, marker):
        for idx, row in plot_df.iterrows():
            length = row['sid_length']
            size = row['codebook_size']
            
            rgb = len_to_color[length]
            alpha = size_to_alpha[size]
            rgba = (*rgb, alpha)
            
            target_ax.scatter(
                length, 
                row[metric_col], 
                color=rgba, 
                marker=marker,
                s=250, 
                edgecolor='white', 
                linewidth=1.0,
                zorder=3
            )

    plot_points(ax_top, "FG/memorization", "^")
    plot_points(ax_bot, "FG/generalization", "o")

    # Limits & Broken Axis Logic
    mem_min, mem_max = plot_df["FG/memorization"].min(), plot_df["FG/memorization"].max()
    gen_min, gen_max = plot_df["FG/generalization"].min(), plot_df["FG/generalization"].max()
    mem_pad = (mem_max - mem_min) * 0.1
    gen_pad = (gen_max - gen_min) * 0.1
    
    ax_top.set_ylim(mem_min - mem_pad, mem_max + mem_pad)
    ax_bot.set_ylim(gen_min - gen_pad, gen_max + gen_pad)

    ax_top.spines['bottom'].set_visible(False)
    ax_bot.spines['top'].set_visible(False)
    ax_top.xaxis.tick_top()
    ax_top.tick_params(labeltop=False)
    ax_bot.xaxis.tick_bottom()
    
    d = .015
    kwargs = dict(transform=ax_top.transAxes, color='k', clip_on=False)
    ax_top.plot((-d, +d), (-d, +d), **kwargs)
    ax_top.plot((1 - d, 1 + d), (-d, +d), **kwargs)
    kwargs.update(transform=ax_bot.transAxes)
    ax_bot.plot((-d, +d), (1 - d, 1 + d), **kwargs)
    ax_bot.plot((1 - d, 1 + d), (1 - d, 1 + d), **kwargs)

    ax_bot.set_xlabel("Codebook Length (Depth)", fontsize=16, labelpad=10)
    ax_bot.set_xticks(unique_lengths)
    ax_bot.set_xticklabels(unique_lengths)
    
    ax_top.text(0.02, 0.9, "Memorization", transform=ax_top.transAxes, fontsize=14, fontweight='bold', color='#444444')
    ax_bot.text(0.02, 0.9, "Generalization", transform=ax_bot.transAxes, fontsize=14, fontweight='bold', color='#444444')

    legend_shapes = [
        Line2D([0], [0], marker='^', color='w', label='Memorization', markerfacecolor='gray', markersize=10),
        Line2D([0], [0], marker='o', color='w', label='Generalization', markerfacecolor='gray', markersize=10)
    ]
    ax_top.legend(handles=legend_shapes, loc='lower right', frameon=True)

    # --- Colorbars (Tight Spacing) ---
    # wspace=0.0 ensures no whitespace between bars
    gs_cb = gs[1].subgridspec(1, 4, wspace=0.0) 
    norm = mcolors.Normalize(vmin=0, vmax=len(unique_sizes)-1)

    for i, length in enumerate(unique_lengths):
        ax_cb = fig.add_subplot(gs_cb[0, i])
        base_color = len_to_color[length]
        
        c_start = (*base_color, 0.3)
        c_end = (*base_color, 1.0)
        cmap = mcolors.LinearSegmentedColormap.from_list(f"len_{i}", [c_start, c_end])
        
        cb = plt.colorbar(plt.cm.ScalarMappable(cmap=cmap, norm=norm), cax=ax_cb)
        
        # Labeling (only title for length)
        ax_cb.set_title(f"L{length}", fontsize=10, pad=5)
        
        # Remove ticks/borders to make them look seamless
        cb.outline.set_visible(False)
        
        if i == len(unique_lengths) - 1:
            cb.set_ticks(np.arange(len(unique_sizes)))
            cb.set_ticklabels(unique_sizes)
            ax_cb.set_ylabel('Codebook Size', rotation=270, labelpad=15)
        else:
            ax_cb.set_yticks([])
            ax_cb.set_xticks([])
    plt.savefig('outputs/test_summary.pdf', bbox_inches='tight')
    plt.show()

# Run with actual dataframe
if 'test_summary' in globals():
    plot_test_summary_final_corrected(test_summary)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.colors as mcolors
from mpl_toolkits.axes_grid1 import make_axes_locatable

# ------------------------------------------------------------------
# Style Configuration
# ------------------------------------------------------------------
sns.set_style("whitegrid")
plt.rcParams.update({
    "font.family": "serif",
    "font.serif": ["Times New Roman"],
    "font.size": 14,
    "axes.labelsize": 16,
    "axes.titlesize": 18,
    "xtick.labelsize": 14,
    "ytick.labelsize": 14,
    "legend.fontsize": 12,
    "axes.linewidth": 1.5,
    "grid.linewidth": 0.5,
    "grid.alpha": 0.6,
    "pdf.fonttype": 42,
    "ps.fonttype": 42,
})

def plot_training_dynamics_fixed(val_df, smoothing_alpha=0.3):
    # 1. Data Prep & CORRECTION
    df = val_df.copy()
    
    df['codebook_size'] = df['sid'].apply(lambda x: int(x.split('x')[0]))
    df['sid_length'] = df['sid'].apply(lambda x: int(x.split('x')[1]) + 1) # Add 1 here
    
    metrics = ["FG/memorization", "FG/generalization"]
    for m in metrics:
        df[m] = df.groupby("sid")[m].transform(lambda x: x.ewm(alpha=smoothing_alpha).mean())

    unique_lengths = sorted(df['sid_length'].unique())
    unique_sizes = sorted(df['codebook_size'].unique())
    
    colors = sns.color_palette("colorblind", n_colors=len(unique_lengths))
    len_to_color = dict(zip(unique_lengths, colors))
    
    size_alphas = np.linspace(0.3, 1.0, len(unique_sizes))
    size_to_alpha = dict(zip(unique_sizes, size_alphas))

    # 2. Setup Figure: 2 cols (metrics) + 1 narrow col for colorbars
    # width_ratios: [1, 1, 0.05] makes the colorbar column very thin
    fig = plt.figure(figsize=(18, 8))
    gs = fig.add_gridspec(2, 3, width_ratios=[1, 1, 0.05], height_ratios=[3, 1], hspace=0.08, wspace=0.15)
    
    # 3. Plotting Loop (Metrics)
    for col_idx, metric in enumerate(metrics):
        ax_top = fig.add_subplot(gs[0, col_idx])
        ax_bot = fig.add_subplot(gs[1, col_idx], sharex=ax_top)
        axes_col = [ax_top, ax_bot]
        
        for sid, group in df.groupby("sid"):
            length = group['sid_length'].iloc[0]
            size = group['codebook_size'].iloc[0]
            
            base_rgb = len_to_color[length]
            alpha = size_to_alpha[size]
            rgba_color = (*base_rgb, alpha)
            
            for ax in axes_col:
                ax.plot(group['epoch'], group[metric], color=rgba_color, linewidth=2.5)

        # 4. Configure Broken Axis Limits
        y_max = df[metric].max() * 1.02
        
        if "memorization" in metric:
            cut_low, cut_high = 0.12, 0.20
            ax_top.set_ylim(cut_high, y_max)
            ax_bot.set_ylim(0, cut_low)
        else: # Generalization
            cut_low, cut_high = 0.013, 0.018
            ax_top.set_ylim(cut_high, y_max)
            ax_bot.set_ylim(0.005, cut_low)

        # 5. Styling the Break
        ax_top.spines['bottom'].set_visible(False)
        ax_bot.spines['top'].set_visible(False)
        ax_top.xaxis.tick_top()
        ax_top.tick_params(labeltop=False)
        ax_bot.xaxis.tick_bottom()

        d = .015 
        kwargs = dict(transform=ax_top.transAxes, color='k', clip_on=False)
        ax_top.plot((-d, +d), (-d, +d), **kwargs)
        ax_top.plot((1 - d, 1 + d), (-d, +d), **kwargs)
        kwargs.update(transform=ax_bot.transAxes)
        ax_bot.plot((-d, +d), (1 - d, 1 + d), **kwargs)
        ax_bot.plot((1 - d, 1 + d), (1 - d, 1 + d), **kwargs)
        
        ax_top.set_title(metric.split("/")[-1].capitalize())
        if col_idx == 0:
            ax_top.set_ylabel("Metric Value")
            
        ax_top.grid(True, which='major', axis='y', linestyle='--', alpha=0.5)
        ax_bot.grid(True, which='major', axis='y', linestyle='--', alpha=0.5)

    # Shared X Label
    fig.text(0.45, 0.04, 'Epoch', ha='center', fontsize=16)

    # 6. Tight Colorbars (No whitespace)
    # Using gs[:, 2] to span full height
    # wspace=0.0 forces them together
    gs_cb = gs[:, 2].subgridspec(1, 4, wspace=0.0) 
    
    norm = mcolors.Normalize(vmin=0, vmax=len(unique_sizes)-1)
    
    for i, length in enumerate(unique_lengths):
        ax_cb = fig.add_subplot(gs_cb[0, i])
        
        base_color = len_to_color[length]
        c_start = (*base_color, 0.3)
        c_end = (*base_color, 1.0)
        cmap = mcolors.LinearSegmentedColormap.from_list(f"len_{i}", [c_start, c_end])
        
        cb = plt.colorbar(
            plt.cm.ScalarMappable(cmap=cmap, norm=norm),
            cax=ax_cb,
            orientation='vertical'
        )
        
        # Labeling
        ax_cb.set_title(f"L{length}", fontsize=10, pad=5)
        
        # Remove outline for seamless look
        cb.outline.set_visible(False)
        
        if i == len(unique_lengths) - 1:
            cb.set_ticks(np.arange(len(unique_sizes)))
            cb.set_ticklabels(unique_sizes)
            ax_cb.set_ylabel('Codebook Size (Alpha)', rotation=270, labelpad=15)
        else:
            ax_cb.set_yticks([])
            ax_cb.set_xticks([])
    
    plt.savefig('outputs/training_dynamics.pdf', bbox_inches='tight')
    plt.show()

# Run with your dataframe
if 'val_dynamics' in globals():
    plot_training_dynamics_fixed(val_dynamics, smoothing_alpha=0.1)
```

```python
# ==========================================
# Label computation functions (reusable for multi-SID analysis)
# ==========================================

def get_item_case_labels(test_item_seqs, fine_grained_evaluator):
    """Get item-level labels for each test case."""
    item_case_labels = {}
    for idx, item_seq in enumerate(test_item_seqs):
        item_case_labels[idx] = fine_grained_evaluator.get_case_labels(item_seq)
    return item_case_labels


def build_prefix_evaluators(train_item_seqs, tokenizer, prefix_lengths, max_hop):
    """Build prefix evaluators for each prefix length."""
    evaluators = {}
    for plen in prefix_lengths:
        evaluators[plen] = PrefixGramMemorizationEvaluator(
            train_item_seqs=train_item_seqs,
            tokenizer=tokenizer,
            prefix_length=plen,
            max_hop=max_hop,
        )
    return evaluators


def get_token_case_labels(test_item_seqs, prefix_evaluators, prefix_lengths):
    """Get token-level labels for each test case."""
    token_case_labels = {}
    for idx, item_seq in tqdm(enumerate(test_item_seqs), total=len(test_item_seqs), desc="Token labels"):
        token_case_labels[idx] = {}
        for plen in prefix_lengths:
            token_case_labels[idx][plen] = prefix_evaluators[plen].get_case_labels(item_seq)
    return token_case_labels


def get_token_category(token_labels_dict, prefix_lengths):
    """Longest memorized prefix, or 'unseen'. Token categories are exclusive."""
    for pl in sorted(prefix_lengths, reverse=True):
        labels = token_labels_dict.get(pl, set())
        if labels and 'unseen' not in labels:
            return f'{pl}-gram'
    return 'unseen'


def is_item_generalization(item_labels):
    """True if instance is item-level generalization (excl. memorization)."""
    if 'memorization' in item_labels:
        return False
    return True
    # sub_cats = ['contractivity', 'symmetry', 'transitivity', '2nd-symmetry', 'novelty']
    # return any(label.split('_')[0] in sub_cats for label in item_labels)


def compute_conversion_rates(item_case_labels, token_case_labels, prefix_lengths):
    """
    Conversion rate: % of item-generalization instances that convert to token memorization
    at each prefix length. Token categories are exclusive (1-gram, 2-gram, ..., unseen).
    """
    gen_indices = [idx for idx in item_case_labels if is_item_generalization(item_case_labels[idx])]
    n_gen = len(gen_indices)
    if n_gen == 0:
        return {plen: 0.0 for plen in prefix_lengths}

    rates = {}
    for plen in prefix_lengths:
        token_cat = f'{plen}-gram'
        n_convert = sum(
            1 for idx in gen_indices
            if get_token_category(token_case_labels[idx], prefix_lengths) == token_cat
        )
        rates[plen] = 100 * n_convert / n_gen
    return rates
```

```python
# ==========================================
# Collision Rate & Conversion Rate Statistics
# ==========================================

def compute_prefix_stats(item2sem_ids):
    """
    Compute prefix-level statistics from semantic IDs.
    
    Returns:
        dict with: n_items, n_digits, prefix_collisions, collision_rate, 
                   unique_prefixes, avg_items_per_prefix, branching_factor
    """
    if not item2sem_ids:
        return None
    
    # Handle both tuple and list values (JSON loads as list)
    def to_tuple(sid):
        if isinstance(sid, (list, tuple)):
            return tuple(int(x) for x in sid)
        return (sid,)
    
    n_items = len(item2sem_ids)
    n_digits = len(next(iter(item2sem_ids.values())))
    
    stats = {
        'n_items': n_items,
        'n_digits': n_digits,
        'prefix_collisions': {},      # prefix_len -> count of items in colliding groups
        'collision_rate': {},         # prefix_len -> % of items that share prefix with others
        'unique_prefixes': {},       # prefix_len -> number of unique prefixes
        'avg_items_per_prefix': {},  # prefix_len -> avg
        'max_collision_size': {},    # prefix_len -> max items sharing same prefix
    }
    
    for prefix_len in range(1, n_digits + 1):
        prefix2items = defaultdict(list)
        for item, sid in item2sem_ids.items():
            t = to_tuple(sid)
            prefix = t[:prefix_len]
            prefix2items[prefix].append(item)
        
        n_unique = len(prefix2items)
        stats['unique_prefixes'][prefix_len] = n_unique
        
        # Collision: items that share a prefix with at least one other item
        n_colliding = sum(len(items) for items in prefix2items.values() if len(items) > 1)
        stats['prefix_collisions'][prefix_len] = n_colliding
        stats['collision_rate'][prefix_len] = 100 * n_colliding / n_items if n_items else 0
        
        # Avg items per prefix
        stats['avg_items_per_prefix'][prefix_len] = n_items / n_unique if n_unique else 0
        
        # Max collision size
        stats['max_collision_size'][prefix_len] = max(len(items) for items in prefix2items.values()) if prefix2items else 0
    
    return stats

# Compute stats for each SID config
sid_stats = {}
for config_name, item2sem_ids in sid_data.items():
    sid_stats[config_name] = compute_prefix_stats(item2sem_ids)

# Summary table
print("=== Collision Rate by SID Config (at each prefix length) ===\n")
stats_df = []
for config_name, stats in sid_stats.items():
    for plen in stats['unique_prefixes']:
        stats_df.append({
            'config': config_name,
            'prefix_len': plen,
            'collision_rate_pct': stats['collision_rate'][plen],
            'unique_prefixes': stats['unique_prefixes'][plen],
            'avg_items_per_prefix': stats['avg_items_per_prefix'][plen],
            'max_collision_size': stats['max_collision_size'][plen],
        })
stats_df = pd.DataFrame(stats_df)
# stats_df
```

```python
# ==========================================
# High-Level SID Config Summary
# ==========================================
# Semantic part = all digits except last (last digit = identifier).
# Collision = multiple items share same semantic part but have different identifier tokens.

def parse_config(config_name):
    """Parse 'LxS' (e.g. '4x64') -> (codebook_length, codebook_size)."""
    parts = config_name.split('x')
    if len(parts) == 2:
        return int(parts[0]), int(parts[1])
    return None, None

summary_rows = []
for config_name, stats in sid_stats.items():
    if stats is None:
        continue
    cb_len, cb_size = parse_config(config_name)
    if cb_len is None:
        cb_len = stats['n_digits']
        cb_size = max(max(sid) for sid in sid_data[config_name].values()) + 1 if sid_data[config_name] else 0

    n_items = stats['n_items']
    n_digits = stats['n_digits']
    # Semantic part = all digits except last (last is identifier)
    semantic_len = n_digits - 1
    expressivity = cb_size ** semantic_len
    n_unique_tuples = stats['unique_prefixes'][semantic_len]
    max_collision = stats['max_collision_size'][semantic_len]
    avg_items_per_tuple = stats['avg_items_per_prefix'][semantic_len]
    collision_rate_pct = stats['collision_rate'][semantic_len]

    summary_rows.append({
        'config': config_name,
        'expressivity': expressivity,
        'n_unique_tuples': n_unique_tuples,
        'n_items': n_items,
        'max_collision': max_collision,
        'avg_items_per_tuple': round(avg_items_per_tuple, 2),
        'collision_rate_pct': round(collision_rate_pct, 1),
    })

summary_df = pd.DataFrame(summary_rows)
print("=== SID Config High-Level Summary ===\n")
print(summary_df.to_string(index=False))
```

```python
# ==========================================
# Global Item Generalization → Token Memorization Conversion Rate
# ==========================================
# For each SID config: % of item-generalization instances that convert to each
# token category. Prefix length 0 = unseen, 1–4 = k-gram memorization.
# Output: table with rows = SID configs, columns = prefix lengths [0,1,2,3,4].

def compute_global_conversion_rates(item_case_labels, token_case_labels, prefix_lengths):
    """
    Global conversion rate: % of item-generalization instances that convert to each token category.
    Returns dict with keys 0 (unseen), 1 (1-gram), 2 (2-gram), 3 (3-gram), 4 (4-gram).
    """
    max_k = max(prefix_lengths)
    gen_indices = [idx for idx in item_case_labels if is_item_generalization(item_case_labels[idx])]
    n_gen = len(gen_indices)
    if n_gen == 0:
        return {k: 0.0 for k in range(max_k + 1)}

    rates = {k: 0.0 for k in range(max_k + 1)}
    for idx in gen_indices:
        token_cat = get_token_category(token_case_labels[idx], prefix_lengths)
        if token_cat == 'unseen':
            rates[0] += 1
        else:
            k = int(token_cat.split('-')[0])
            if k in rates:
                rates[k] += 1

    for k in rates:
        rates[k] = 100.0 * rates[k] / n_gen
    return rates


def resolve_sem_ids_path(base_path):
    """Try path as-is, then with _faiss before .sem_ids."""
    if os.path.exists(base_path):
        return base_path
    base, ext = os.path.splitext(base_path)
    alt = base + '_faiss' + ext
    return alt if os.path.exists(alt) else base_path


def parse_sid_config(config_name):
    """Parse 'LxS' (e.g. '4x64') -> (n_codebooks, codebook_size)."""
    parts = config_name.split('x')
    if len(parts) == 2:
        return int(parts[0]), int(parts[1])
    return None, None


# Compute conversion rates for each SID config
PREFIX_COLS = [0, 1, 2, 3, 4, 5]
global_conv_rows = []

for config_name, path in tqdm(SID_PATHS.items(), desc="SID configs"):
    resolved = resolve_sem_ids_path(path if os.path.isabs(path) else os.path.join(os.getcwd(), path))
    if not os.path.exists(resolved):
        print(f"Skip {config_name}: path not found")
        continue

    n_cb, cb_size = parse_sid_config(config_name)
    if n_cb is None:
        continue

    try:
        config_dict = {
            'category': CATEGORY,
            'logging': False,
            'sem_ids_path': resolved,
            'rq_n_codebooks': n_cb,
            'rq_codebook_size': cb_size,
        }
        pipe = Pipeline(model_name='TIGER', dataset_name=DATASET_NAME, config_dict=config_dict)
        tok = pipe.tokenizer
        # Use n_codebooks, NOT tok.n_digit. TIGER's n_digit = n_codebooks + 1 (conflict resolution).
        # Items have n_codebooks tokens; using n_digit creates extra evaluator → 100% in wrong column.
        prefix_lengths = list(range(1, n_cb + 2))
        print(prefix_lengths)

        train_seqs = pipe.split_datasets['train']['item_seq']
        peval = build_prefix_evaluators(train_seqs, tok, prefix_lengths, MAX_HOP)
        tlabels = get_token_case_labels(test_item_seqs, peval, prefix_lengths)

        rates = compute_global_conversion_rates(item_case_labels, tlabels, prefix_lengths)
        row = {'config': config_name}
        for k in PREFIX_COLS:
            row[k] = rates.get(k, 0.0)
        global_conv_rows.append(row)
    except Exception as e:
        print(f"Skip {config_name}: {e}")

# Build table: rows = SID configs, columns = prefix lengths [0,1,2,3,4]
df_global_conv = pd.DataFrame(global_conv_rows)
if not df_global_conv.empty:
    df_global_conv = df_global_conv.set_index('config')[PREFIX_COLS]
    print("\nGlobal conversion rate (%): item generalization → token category")
    print("Rows: SID configs | Cols: 0=unseen, 1–4=k-gram memorization")
df_global_conv.round(2)
```
